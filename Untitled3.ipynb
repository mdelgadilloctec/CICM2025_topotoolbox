```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
import shutil

drive_path = '/content/drive/MyDrive'
folder_name = 'curso_CICM2025'
new_folder_path = os.path.join(drive_path, folder_name)

if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)
else:
  print(f"El directorio '{folder_name}' ya existe en Google Drive.")
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
import os
import shutil

# Get the actual uploaded filename and content from the 'uploaded' dictionary
if uploaded:
    source_filename = list(uploaded.keys())[0]
    source_content = uploaded[source_filename]
    source_file_path_in_colab = os.path.join('/content/', source_filename)

    # Explicitly write the content to a file in /content/
    try:
        with open(source_file_path_in_colab, 'wb') as f:
            f.write(source_content)
        print(f"File '{source_filename}' successfully written to '{source_file_path_in_colab}'.")
    except Exception as e:
        print(f"Error writing file to /content/: {e}")
        source_file_path_in_colab = None # Indicate failure

    if source_file_path_in_colab:
        destination_file = os.path.join(new_folder_path, 'dem_sin_depresiones.tif') # Renaming for consistency

        # Move the file to the new folder in Google Drive
        try:
            shutil.move(source_file_path_in_colab, destination_file)
            print(f"File moved from '{source_file_path_in_colab}' to '{destination_file}' successfully.")
            ruta = destination_file # Update 'ruta' to the new location
            print(f"'ruta' variable updated to: {ruta}")
        except FileNotFoundError:
            print(f"Error: Source file '{source_file_path_in_colab}' not found. This should not happen if it was just written.")
        except Exception as e:
            print(f"An error occurred while moving the file: {e}")
else:
    print("Error: No file was uploaded in the previous step (variable 'uploaded' is empty).")
```

```python
import topotoolbox as tt3
import matplotlib.pyplot as plt
dem = tt3.read_tif('/content/drive/MyDrive/curso_CICM2025/dem_sin_depresiones.tif')
dem.info()
```

```python
im = dem.plot(cmap="cubehelix")
plt.colorbar(im)
plt.show()
```
